```python
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(12,8))
width = 0.2

r=np.arange(4)

data_normal=[result[i]['mean_tau_after_ordering'] for i in range(4)]
data_without_context=[result_ignore_context[i]['mean_tau_after_ordering'] for i in range(4)]
data_without_end_order_value=[result_ignore_end_order_value[i]['mean_tau_after_ordering'] for i in range(4)]
data_before_ordering=[result[i]['mean_tau_before_ordering'] for i in range(4)]

plt.bar(r,data_before_ordering , color = "tab:grey",
        width = width, edgecolor = 'black',
        label="Before ordering")

plt.bar(r+width,data_normal , color = "tab:red",
        width = width, edgecolor = 'black',
        label="After ordering")

plt.bar(r+width*2,data_without_end_order_value , color = "tab:green",
        width = width, edgecolor = 'black',
        label="After ordering, without using end order value")

plt.bar(r+width*3,data_without_context , color = "tab:blue",
        width = width, edgecolor = 'black',
        label="After ordering, without using context")

plt.ylabel("Mean kendall tau")
plt.title("Comparing the mean kendall tau before and after ordering")
plt.yticks([i/20 for i in range(21)])
plt.grid(axis='y')
plt.xticks(r + width*1.5,["Semantic filtering level "+str(i) for i in range(4)])
plt.legend()
plt.legend(loc=9, bbox_to_anchor=(0.5, -0.125))
plt.show()
```
