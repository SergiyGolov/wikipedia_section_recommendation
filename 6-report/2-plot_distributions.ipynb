```python
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
%matplotlib inline

def plot_nb_different_categories_by_article(articles_categories_file):
    plt.figure(figsize=(14,8))
    
    article_categories={}
    with open(articles_categories_file) as f:
        for line in f:
            split_line = line.split()
            article_id = int(split_line[0])
            category = split_line[2]
            if article_id not in article_categories.keys():
                article_categories[article_id]=[]
            article_categories[article_id].append(category)
    
    article_category_counter=Counter()
    for article,categories in article_categories.items():
        article_category_counter[len(categories)]+=1
    
    data=[]
    for k,v in article_category_counter.items():
        if k<=25:
            for _ in range(v):
                data.append(k)
            
    plot=sns.histplot(data=data,stat="probability",bins=list(range(1,26)))
    plot.set_xticks(list(range(1,26)))
    
   
    #plot.set_yticks(list(set(counter.values())))
    plot.set_title("Number of categories to which articles belong")
    plot.set_xlabel("Number of categories by article")
    plot.set_ylabel("Proportion")
    plt.show()
```

```python
plot_nb_different_categories_by_article("../data/article_categories_filtered.tsv")
```
