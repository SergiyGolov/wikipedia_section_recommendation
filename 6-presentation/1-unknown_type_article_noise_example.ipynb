```python
categories_pure_after_unknown_type_article_filtering=set()
```

```python
articles_having_type=set()
article_types=read_tuple_list_from_file((int,str),"../data/article_types.tsv")
for article_id,_ in article_types:
    articles_having_type.add(article_id)
```

```python
from collections import Counter
```

```python
category_by_nb_articles=Counter()
with open("../data/gini_articles_scores_0985.json") as f:
    for line in f:
        line_dict = json.loads(line)
        category = line_dict['category']
        if len(line_dict['articles'])>0:
            proportion_unknown_types=len([x for x in line_dict['articles'] if x not in articles_having_type])/len(line_dict['articles'])
            if proportion_unknown_types>0.5:
                category_by_nb_articles[category]+=len([x for x in line_dict['articles'] if x not in articles_having_type])
```

```python
categories_with_articles=set()
```

```python
article_categories=read_tuple_list_from_file((int,str,str),'../data/article_categories_no_unknown_types.tsv')
for _,_,cat in article_categories:
    categories_with_articles.add(cat)
```

```python
for cat,count in category_by_nb_articles.most_common(1000):
    if cat in categories_pure_after_unknown_type_article_filtering and cat in categories_with_articles:
        print(f"{cat} : {count}")
```
